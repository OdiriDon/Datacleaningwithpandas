```python
#First I import my pandas library
#And import my dataset
import pandas as pd
aircrash = pd.read_csv("aircrash1908-2008.csv")
aircrash.head(3)
```

```python
#Previewing our data I noticed that our location has both the city and country together so we have to seperate them
#Using the split method
aircrash[["City", "Country"]]  = aircrash["Location"].str.split(",", expand = True, n = 1)
```

```python
#We now drop the location column and preview it
Aircrash = aircrash.drop("Location", axis = 1)
```
